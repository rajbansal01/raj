```python
#Add a new product

def add_prod():
    clear_output()
    key = input("Enter product id: ")
    if key in products:
        print("Product with this ID is already present in our inventory.\nPlease check the product ID you've entered and try again!")
    else:
        v = {}
        nm = input("Enter name of product: ")
        pr = int(input("Enter price of product: "))
        mv = input("Enter mass/volume of product: ")
        r_qty = int(input("Enter quantity: "))
        value = {"Name": nm, "Price": pr, "Mass/Vol": mv, "Available Quantity": r_qty}
        products[key] = value
        print("Product added successfully ")
    q = input("Do you want to add more products?(Y/N): ")
    if (q == "y" or q == "Y"):
        add_prod()
    elif (q == "n" or q == "N"):
        main()
    else:
        print("Invalid Choice! Returning to main menu in 3 seconds...")
        time.sleep(3)
        main()
    
```

```python
# Update a Product

def update_prod():
    clear_output()
    uid = input("Enter the id of product you want to update: ")
    if uid in products:
        print("Product ID: ", uid)
        print("Name: ", products[uid]["Name"])
        products[uid]["Price"] = int(input("Enter the price: "))
        products[uid]["Mass/Vol"] = input("Enter mass/volume of product: ")
        products[uid]["Available Quantity"] = int(input("Enter total quantity available: "))
    else:
        print("Product with this ID is not from our inventory.\nPlease check the product ID you've entered and try again!")
    
    q = input("Do you want to update more products in your inventory?(Y/N): ")
    if (q == "y" or q == "Y"):
        update_prod()
    elif (q == "n" or q == "N"):
        main()
    else:
        print("Invalid Choice! Returning to main menu in 3 seconds...")
        time.sleep(3)
        main()
```

```python
# Remove a Product

def rem_prod():
    clear_output()
    d_pro = input("Enter the ID of product You want to remove: ")
    if d_pro in products:
        del products[d_pro]
        print("Product removed successfully!")
    else:
        print("Product with this ID is not from our inventory.\nPlease check the product ID you've entered and try again!")
    
    q = input("Do you want to remove more products from your inventory?(Y/N): ")
    if (q == "y" or q == "Y"):
        rem_prod()
    elif (q == "n" or q == "N"):
        main()
    else:
        print("Invalid Choice! Returning to main menu in 3 seconds...")
        time.sleep(3)
        main()
```

```python
# Add product to your cart

def add_cart():
    clear_output()
    ui_id = input("Write the product id of the item u want to purchase: ")
    if ui_id in products:
        i = ui_id
        ui_qty = int(input("How much quantity do you want? "))
        if (ui_qty < products[i]["Available Quantity"]):  
            if(len(cart) == 0):
                c = 1
                lst = [c,i, products[i]["Name"], products[i]["Price"], products[i]["Mass/Vol"], ui_qty]
                cart.append(lst)
                c += 1
                print("Products added to cart!")
            else:
                temp = []
                for j in range(len(cart)):
                    temp.append(cart[j][1])
                if i in temp:
                    j = temp.index(i)
                    cart[j][5] = cart[j][5] + ui_qty
                else:
                    lst = [len(cart)+1,i, products[i]["Name"], products[i]["Price"], products[i]["Mass/Vol"], ui_qty]
                    cart.append(lst)
                    print("Products added to cart!")
        else:
            print("Sorry for your inconvenience, but we don't have enough quantity for the item you require!\nPlease check the list of products from the main menu.")
    else:
        print("Product with this ID is not available in our inventory.\nPlease check the product ID you've entered and try again!")
    
    q = input("Do you want to add more products to your cart?(Y/N): ")
    if (q == "y" or q == "Y"):
        add_cart()
    elif (q == "n" or q == "N"):
        main()
    else:
        print("Invalid Choice! Returning to main menu in 3 seconds...")
        time.sleep(3)
        main()     
```

```python
# Update Sales in document

def update_sales():
    filesize = os.path.getsize("Total_Sales.txt")
    if (filesize == 0):
        t_sales = {}
        for i in range(len(cart)):
            t_sales[cart[i][1]] = {"Name": cart[i][2], "Price": cart[i][3], "Mass/Vol": cart[i][4], "Quantity Sold": cart[i][5]}
        
    else:
        ft = open("Total_Sales.txt", 'r')
        jt = ft.read()
        ft.close()
        t_sales = json.loads(jt)
        ts_keys = list(t_sales.keys())
        for i in range(len(cart)):
            if cart[i][1] in ts_keys:
                t_sales[cart[i][1]]["Quantity Sold"] = t_sales[cart[i][1]]["Quantity Sold"] + cart[i][5]
            else:
                t_sales[cart[i][1]] = {"Name": cart[i][2], "Price": cart[i][3], "Mass/Vol": cart[i][4], "Quantity Sold": cart[i][5]}
    temp = {}
    for i in sorted (t_sales) :
        temp[i] = t_sales[i]
    t_sales = temp
    jt = json.dumps(t_sales)                
    ft = open("Total_Sales.txt", 'w')
    ft.write(jt)
    ft.close()
```
